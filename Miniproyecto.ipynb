## Ejercicio1 - El triangulo de Shierpinski

```python
import random
import math
import matplotlib.pyplot as plt
```

```python
#Puntos donde empieza el fractal
p1, p2, p3 = 0.20, 0.30, 0.50
#Numero de iteraciones
it = 100000
x, y = [],[]

def transf(x, y, prob):
    
    if(prob <= p1):
        x = x/2
        y = y/2
    elif(prob <= (p1+p2) and prob > p1):
        x = x/2 + 0.5
        y = y/2
    elif(prob <= (p1+p2+p3) and prob > (p1+p2)):
        x = x/2 + 0.25
        y = y/2 + 0.5
    return x, y
```

```python
pointx = random.random()
pointy = random.random()
```

```python
for iteracion in range(0, it):
    prob = random.random()
    px, py = transf(pointx, pointy, prob)
    x.append(px)
    y.append(py)
    pointx = px
    pointy = py
```

```python
plt.scatter(x, y, s = 0.3, edgecolor = 'yellow')
plt.show()
```

## Ejercicio 2 - El helecho de Barnsley

```python
from random import random
import matplotlib.pyplot as plt
```

```python
def funcion_1 (x, y):
    return (x * 0.85 + y * 0.04 + 0.0, x * (-0.04) + y * 0.85 + 1.6)

def funcion_2 (x, y):
    return (x * (-0.15) + y * 0.28 + 0.0, x * (0.26) + y * 0.24 + 0.44)

def funcion_3 (x, y):
    return (x * 0.2 + y * (-0.26) + 0.0, x * (0.23) + y * 0.22 + 1.6)

def funcion_4 (x, y):
    return (x * 0.0 + y * 0.0, x * 0.0 + y * 0.16)
```

```python
x, y = [0], [0]
N = 100000

for i in range(1, N):
    random_number = random()

    a, b = 0, 0
    if random_number < 0.85:
        a, b = funcion_1(x[i - 1], y[i - 1])
    elif random_number < 0.85 + 0.07:
        a, b = funcion_2(x[i - 1], y[i - 1])
    elif random_number < 0.85 + 0.07 + 0.07:
        a, b = funcion_3(x[i - 1], y[i - 1])
    else:
        a, b = funcion_4(x[i - 1], y[i - 1])

    x.append(a)
    y.append(b)
```

```python
plt.scatter(x, y, s = 0.2, edgecolor = 'green')
plt.show()
```

## Ejercicio 3 - Análisis de pseudorandoms

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import random
```

```python
def pseudorandom(a, m):
    def inner_pseudo(xo, n):
        items = list(range(n))
        items[0] = xo

        for i in range(1, n):
            items[i] = (a * items[i - 1]) % m
        return [i / float(m) for i in items]
    return inner_pseudo
```

```python
RANDOM_1 = (5**5, 2**35 - 1)
RANDOM_2 = (7**5, 2**31 - 1)
```

```python
pseudorandom_1 = pseudorandom(*RANDOM_1)
pseudorandom_2 = pseudorandom(*RANDOM_2)
```

## Ejercicio 5 - Integral bidimensional

```python
import math
```

$$ \theta=\int_{0}^{\infty}\int_{0}^{x}e^{-(x+y)}dy dx = \int_{0}^{\infty} (e^{-x} - e^{-2x}) dx = \int_{0}^{1} \frac{e^{-(\frac{1}{x} - 1)} - e^{-2(\frac{1}{x} - 1)}}{x^{2}} dx $$ 



```python
def pseudorandom(a, m):
    def inner_pseudo(xo, n):
        items = list(range(n))
        items[0] = xo

        for i in range(1, n):
            items[i] = (a * items[i - 1]) % m
        return [i / float(m) for i in items]
    return inner_pseudo

xo = 200
```

```python
g = lambda x: ( ( (math.e**(-((1/x) - 1))) - (math.e**(-2*((1/x) - 1))) ) / (x**2) )
```

### 100 iteraciones

```python
N = 100
```

```python
RANDOM_2 = (7**5, 2**31 - 1)
pseudorandom = pseudorandom(*RANDOM_2)

integral = pseudorandom(xo, N)
```

```python
a_evaluar = [g(i) for i in integral]
sum(a_evaluar) / len(a_evaluar)
```
